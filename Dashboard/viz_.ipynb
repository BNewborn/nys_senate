```python
years = list(set(monthly_topics.year))
months = list(set(monthly_topics.month))
topics = list(set(monthly_topics.topic_desc))
print(topics)
```

```python
pct_adopted_month = {}
list_of_dicts = []
indx = 0
exclude_topics = ['high school athletics honors','honoring firefighters','honoring resolutions','scouts']
for year in years:
    for month in months:
        indx += 1
        
        mini_group = monthly_topics[(monthly_topics['year']==year)\
                       &(monthly_topics['month']==month)&(monthly_topics['adopted']==True)
                        &(~monthly_topics.topic_desc.isin(exclude_topics))]
        
        month_year = dt.date(year=year,month=month,day=1)
        
        #total adopted bills in a month/year
        if mini_group['bills'].sum()==0:
            tot_adopted = 1
        else:
            tot_adopted = mini_group['bills'].sum()
            
    
        for topic in topics:
            
            try:
                top_adopted = mini_group[mini_group['topic_desc']==topic]['bills'].values[0]
            except:
                top_adopted = 0 

            pct_adopted = round(100 * top_adopted / tot_adopted ,2)
                

            pct_adopted_month[indx] = {'topic_desc': topic 
                             ,'month_year': month_year
                             , 'adopted_pct': pct_adopted}
            
            list_of_dicts.append(pct_adopted_month[indx])
bill_adopted_by_month = pd.DataFrame(list_of_dicts)
bill_adopted_by_month['month_year'] = pd.to_datetime(bill_adopted_by_month['month_year'])
bill_adopted_by_month
```
