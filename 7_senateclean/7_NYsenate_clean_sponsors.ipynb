```python
new_list = []
counter = 0
for idx,row in data_sponsor.iterrows():
#     pprint(row)
    counter += 1
#     if counter > 15:
#         break
    if counter % 20000 == 0:
#         clear_output()
        print(f"counter: {counter}")
        
#     pprint(row)
    
    if ( row['sponsor'] == None):
        amend_info = { '_id':row['_id']
                          ,'chamber': ''
                          , 'distictCode':np.NaN
                         ,'fullName': ''
                         , 'sessionYear': np.NaN
                         ,'shortName': ''
                         }
    elif (row['sponsor']['member'] is None):
        if row['sponsor']['budget']:
            amend_info = { '_id':row['_id']
                          ,'chamber': 'budget'
                          , 'distictCode': np.NaN
                         ,'fullName':'budget'
                         , 'sessionYear': np.NaN
                         ,'shortName': 'budget'
                         }
        elif row['sponsor']['rules']:
            amend_info = { '_id':row['_id']
                          ,'chamber': 'rules'
                          , 'distictCode':np.NaN
                         ,'fullName': 'rules'
                         , 'sessionYear': np.NaN
                         ,'shortName': 'rules'
                         }
            

    else:

        member = row['sponsor']['member']
        amend_info = { '_id':row['_id']
                      ,'chamber': member['chamber']
                      , 'distictCode': int(member['districtCode'])
                     ,'fullName': member['fullName']
                     , 'sessionYear': int(member['sessionYear'])
                     ,'shortName': member['shortName']
                     }

    new_list.append(amend_info)
    
    
data_sponsors = pd.DataFrame(new_list)
data_sponsors.sample(10)
```

```python
pd.to_pickle(data_sponsors,"data_sponsors")
```
