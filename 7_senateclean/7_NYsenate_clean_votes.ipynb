```python
def get_rando():
    rando = np.random.randint(90000,100000)
    print(f"rando: {rando}")
    pprint(data_votes.iloc[rando]['votes'])
#     print(data_vetoMessages.iloc[rando]['session'])
get_rando()
```

```python
new_list = []
counter = 0
for idx,row in data_votes.iterrows():
    counter += 1
    
#     if counter > 15:
#         break

    if counter % 20000 == 0:
        print(f"counter: {counter}")
    
    if row['votes']['size'] > 0 and row['votes']['items'][0]['voteType']=='FLOOR':

#         pprint(row['votes']['items'])

        details = row['votes']['items'][0]
#         pprint(details['memberVotes']['items'].keys())
        
        if 'AYE' not in details['memberVotes']['items'].keys() :
            aye_votes = 0
        else:
            aye_votes = details['memberVotes']['items']['AYE']['size']
            
        if 'NAY' not in details['memberVotes']['items'].keys() :
            nay_votes = 0
        else:
            nay_votes = details['memberVotes']['items']['NAY']['size']
        
        if 'EXC' not in details['memberVotes']['items'].keys():
            excused_votes = 0
        else:
            excused_votes = details['memberVotes']['items']['EXC']['size']
            
        vote_info = {
                    'aye_floor_votes' : aye_votes
                    ,'nay_floor_votes': nay_votes
                    ,'other_floor_votes': excused_votes
                    ,'basePrintNoStr':row['basePrintNoStr']
                    }
#         break
    else:
        vote_info = {
                    'aye_floor_votes' : np.NaN
                    ,'nay_floor_votes': np.NaN
                    ,'other_floor_votes': np.NaN
                    ,'basePrintNoStr':row['basePrintNoStr']
                    }
    

    new_list.append(vote_info)
# print(f"Bills with Veto Messages: {counter}")    
    
data_vote_info = pd.DataFrame(new_list)
data_vote_info.sample(10)
```
