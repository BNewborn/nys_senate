```python
def get_rando():
    rando = np.random.randint(90000,100000)
    print(f"rando: {rando}")
    pprint(data_text.iloc[rando]['amendments'])
#     print(data_vetoMessages.iloc[rando]['session'])
get_rando()
```

```python
new_list = []
counter = 0
for idx,row in data_text.iterrows():
    counter += 1
    
#     if counter > 15:
#         break

    if counter % 20000 == 0:
        print(f"counter: {counter}")
#         pprint(row['votes']['items'])
    details = row['amendments']['items']['']
#     print(details.keys())
#     print("*"*50)
    if details['fullText'] =='':
        full_text = 'no_text_available'
    else:
        full_text = details['fullText']
    
    
    text_info = {
                'num_cosponsors' : details['coSponsors']['size']
                ,'full_text': full_text
#                 ,'actClause':details['actClause']
                ,'basePrintNoStr':row['basePrintNoStr']
                }

    

    new_list.append(text_info)
# # print(f"Bills with Veto Messages: {counter}")    
    
data_text_info = pd.DataFrame(new_list)
data_text_info.sample(10)
```
