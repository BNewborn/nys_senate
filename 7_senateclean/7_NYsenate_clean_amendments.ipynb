```python
new_list = []
counter = 0
for idx,row in data_amendments.iterrows():
    counter += 1
#     if counter > 15:
#         break
    if counter % 25000 == 0 :
        print(f"counter {counter}")
    
    items = row['amendments']['items']['']
#     pprint(items)
    ####Same As####
    if items['sameAs']['size'] > 0:
        sameAs = items['sameAs']['items'][0]['basePrintNoStr']
#         print("basePrint SUCCESS")
    else:
        sameAs = ''
    
    amend_info = {'session':items['session']
             ,'basePrintNo':items['printNo']
            ,'printNo_session':items['basePrintNoStr']
            ,'publishDate':items['publishDate']
            ,'lawSection':items['lawSection']
            ,'memo':items['memo']
            ,'fullText':items['fullText']
            ,'coSponsors':items['coSponsors']['size']
            ,'sameAs':sameAs
            }
    
    new_list.append(amend_info)
data_amendments_full = pd.DataFrame(new_list)
data_amendments_full.sample(10)
```

```python
pd.to_pickle(data_amendments_full,"data_text")
```
