```python
data_vetoMessages = data[['_id','vetoMessages','basePrintNoStr']]
data_vetoMessages.sample(10)
```

```python
def get_rando():
    rando = np.random.randint(90000,100000)
    print(f"rando: {rando}")
    pprint(data_vetoMessages.iloc[rando]['vetoMessages'])
#     print(data_vetoMessages.iloc[rando]['session'])
get_rando()
```

```python
new_list = []
counter = 0
for idx,row in data_vetoMessages.iterrows():
    counter += 1
    
#     if counter > 15:
#         break
#     pprint(row)
    if counter % 20000 == 0:
        print(f"counter: {counter}")
    
    if int(row['vetoMessages']['size']) > 0:
        
#         print((row['vetoMessages']['items'][0]['memo']))
        veto_info = {
                    'veto_message' : row['vetoMessages']['items'][0]['memoText']
                    ,'basePrintNoStr': row['basePrintNoStr']
                    }
#         break
    else:
        veto_info = {
                    'veto_message': 'no_veto'
                    ,'basePrintNoStr': row['basePrintNoStr']
                    }
    

    new_list.append(veto_info)
# print(f"Bills with Veto Messages: {counter}")    
    
data_vetoMsgs = pd.DataFrame(new_list)
data_vetoMsgs.sample(10)
```
