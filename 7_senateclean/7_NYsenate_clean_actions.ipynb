```python
data_actions = data[['_id','actions']]
data_actions.iloc[0]['actions']['items'][0]
```

```python
new_list = []
# counter = 0
for idx,row in data_actions.iterrows():
#     counter += 1
#     if counter > 15:
#         break
    items = row['actions']['items']

    for item in items:
        impt_actions = {}
        impt_actions = {'session':item['billId']['session']
                 ,'basePrintNoStr':item['billId']['basePrintNoStr']
                       }
        for key in item:
            if key != 'billId':
                impt_actions[key] = item[key]

        new_list.append(impt_actions)
    
data_actions = pd.DataFrame(new_list)
```

```python
master_data_actions = data_actions[data_actions['sequenceNo']==0].copy()
for i in range(1,data_actions.sequenceNo.max()+1):
    temp_df = data_actions[data_actions['sequenceNo']==i].copy()
    print(f"temp_df size: {temp_df.size}")
    temp_df = temp_df.drop(['session'],axis=1)
    suffix_i = "_"+str(i)
    master_data_actions = master_data_actions.merge(temp_df,how='left'\
                                                   ,on='basePrintNoStr'
                                                   ,suffixes=('',suffix_i))
    if i % 1 == 0:
        print(f"shape: {master_data_actions.shape}")
        print(f"** i: {i} complete **")
```
