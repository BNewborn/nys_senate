```python
senate_master = pd.read_pickle("senate_raw")
```

```python
senate_master.sample()
```

```python
senate_master = senate_master.drop(['actions','additionalSponsors'\
                                    ,'amendmentVersions','amendments','billType'\
                                   ,'calendars','committeeAgendas','milestones'\
                                   ,'pastCommittees','previousVersions','programInfo'\
                                   ,'publishStatusMap','sponsor','status'\
                                   ,'vetoMessages','votes'], axis=1)
```

```python
senate_master.sample()
```

```python
#baseprint
print(f"Size before merges: {senate_master.shape}")
master = senate_master.merge(data_actions\
                             ,left_on='basePrintNoStr'\
                             ,right_on='basePrintNoStr'
                            ,how='left')
print(f"Size after actions merge: {master.shape}")

master = master.merge(data_vetoMessages\
                      ,left_on='basePrintNoStr'\
                      ,right_on='basePrintNoStr'
                     ,how='left')
print(f"Size after vetoMessages merge: {master.shape}")

master = master.merge(data_voteinfo\
                      ,left_on='basePrintNoStr'\
                      ,right_on='basePrintNoStr'
                     ,how='left')
print(f"Size after voteinfo merge: {master.shape}")

master = master.merge(data_text\
                      ,left_on='basePrintNoStr'\
                      ,right_on='basePrintNoStr'
                     ,how='left')
print(f"Size after text merge: {master.shape}")

master = master.merge(data_substitutedBy\
                      ,left_on='basePrintNoStr'\
                      ,right_on='basePrintNoStr'
                     ,how='left')
print(f"Size after subBy merge: {master.shape}")
master = master.drop('substitutedBy',axis=1)

#id
master = master.merge(data_amendmentversions\
                      ,left_on='_id'\
                      ,right_on='_id'
                     ,how='left')
print(f"Size after amendmentversions merge: {master.shape}")

master = master.merge(data_previousVersions\
                      ,left_on='_id'\
                      ,right_on='_id'
                     ,how='left')
print(f"Size after previousVersions merge: {master.shape}")

master = master.merge(data_sponsors\
                      ,left_on='_id'\
                      ,right_on='_id'
                     ,how='left')
print(f"Size after sponsors merge: {master.shape}")






master.sample()





```
