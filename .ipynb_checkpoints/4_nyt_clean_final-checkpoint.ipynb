```python
import pandas as pd
from pprint import pprint
import numpy as np
from IPython.display import clear_output
```

```python
pd.set_option('display.max_columns', 500)
pd.set_option('display.max_colwidth', 100)
```

```python
df = pd.read_pickle("nyt_midway")
```

```python
df.shape
```

```python
df.sample()
```

```python
df_clean = df.drop_duplicates(subset=['snippet','web_url','clean_pubdate'])
```

```python
df_clean.shape
```

Need to extract useful information from following columns (which are dictionaries)
* Byline - 'original' subtract 'by'
* Headline - 'main'

```python
for idx, row in df_clean.iterrows():
    try:
        byline = row['byline']
        print(byline)
        print(f"Type Byline: {type(byline)}")

        author = row['byline']['original']
        print(f"author: {author}")
        df_clean.loc[idx,'author']=author.replace('By ','')
    except:
        df_clean.loc[idx,'author']='no_author'

clear_output()
print('clear output')
```

```python
df_clean.sample()
```

```python
def get_headline(row):
    try:
        headline = row['headline']['main']
        print(f"Headline: {headline}")
        print(f"Type: {type(headline)}")
        row['headline_clean'] = headline
        return row
    except:
        row['headline_clean'] = 'no_headline'
        return row

df_clean = df_clean.apply(get_headline,axis=1)
```

```python
df_clean.sample()
```

```python
df_clean[df_clean['headline_clean']=='no_headline']
```

```python
def editorial_author(row):
    if row['author']=='no_author' and row['new_desk']=='Editorial':
        row['author'] = 'THE EDITORIAL BOARD'
    return row
```

```python
df_clean = df_clean.apply(editorial_author,axis=1)
```

```python
df_clean.loc[(df_clean['author']=='THE EDITORIAL BOARD')]
```

```python
def text_update(row):
    if row['summary']=='NYTimes.com no longer supports \
            Internet Explorer 9 or earlier.\nPlease upgrade your browser.':
        row['summary']=row['snippet']
    return row
```

```python
df_clean = df_clean.apply(text_update,axis=1)
```

```python
df_clean = df_clean[['_id','abstract','headline_clean','keywords','new_desk'\
                     ,'print_page','pub_date','clean_pubdate','year','month','section_name','snippet'\
                    ,'source','type_of_material','web_url','word_count','full_text','summary','author']]
```

```python
pd.to_pickle(df_clean,"nyt_final")
```
