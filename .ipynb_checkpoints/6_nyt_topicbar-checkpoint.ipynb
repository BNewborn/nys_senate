```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
nyt = pd.read_pickle("nyt_master")
```

```python
nyt.sample()
```

```python
#Filter Down to Relevant Fields
nyt_presenting = nyt [['pub_date','headline_clean','new_desk','year','month','snippet','topic_max','topic_desc']]
```

```python
nyt_presenting.sample(5)
```

```python
nyt_presenting[nyt_presenting['topic_max']==14].sort_values(by='pub_date')
```

```python
#Create GroupBy for Presenting
nyt_df = pd.DataFrame(nyt_presenting.groupby(by=['year','month','topic_desc'])['topic_max'].count().sort_values(ascending=False)).reset_index()
```

```python
nyt_df.sort_values(by='year')
```

```python
#Create GroupBy for Presenting
graph = pd.DataFrame(nyt_presenting.groupby(by=['topic_desc'])['topic_max'].count().sort_values(ascending=False)).reset_index()

graph['pct_articles'] = graph['topic_max'].apply(lambda x: 100 * x / graph['topic_max'].sum())
```

```python

plt.style.use('fivethirtyeight')
sns.barplot(y='topic_desc',x='pct_articles',data=graph.head(9))
plt.title("New York Times, State Senate related Articles 2009-2019")
plt.xlabel("Percent of Articles")
plt.ylabel("Topic Descriptions")
```
